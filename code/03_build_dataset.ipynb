# 03 — Final Dataset Construction (NUTS2 Panel)

## Objective

Merge economic and climate data to build a panel dataset at the NUTS2 level (2010–2023).

---

## Data

- Tourism (Eurostat)  
- GDP (Eurostat)  
- Population (Eurostat)  
- Climate (ERA5, processed in Notebook 02)

---

## Method

Datasets are merged using `(region, year)`:

- Inner join for economic variables  
- Left join for climate data  

Data is then cleaned:
- Keep years 2010–2023  
- Drop missing climate observations  

---

## Output

Final dataset:

- ~273 regions  
- ~3,200 observations  

Saved as:

`data/processed/final_dataset_nuts2_2010_2023.csv`

---

```python
import pandas as pd
import numpy as np
import eurostat
```

```python
tourism = pd.read_csv("../data/raw/tourism.csv")
gdp = pd.read_csv("../data/raw/gdp.csv")
pop = pd.read_csv("../data/raw/population.csv")

climate = pd.read_csv("../data/raw/climate_nuts2.csv")
```

```python
df = tourism.copy()

df = df.merge(gdp, on=["region","year"], how="inner")
df = df.merge(pop, on=["region","year"], how="inner")
df = df.merge(climate, on=["region","year"], how="left")
```

```python
print("\n=== MERGE CHECK ===")
print("Shape:", df.shape)

missing = df[["temperature","precipitation"]].isna().mean()
print("Missing ratio:")
print(missing)
```

```python
df = df[df["year"].between(2010, 2023)]

df = df.sort_values(["region","year"]).reset_index(drop=True)

df_clean = df.dropna(subset=["temperature","precipitation"])
```

```python
df_clean.to_csv("../data/processed/final_dataset_nuts2_2010_2023.csv", index=False)

print("Saved:", df_clean.shape)
print("Regions:", df_clean["region"].nunique())
```
